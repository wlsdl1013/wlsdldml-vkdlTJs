# seaborn
* matplotlib의 상위 호환 데이터 시각화를 위한 라이브러리
* 데이터프레임으로 다양한 통계 지표를 낼 수 있는 시각화 차트를 제공하기 때문에 데이터 분석에 활발히 사용되고 있는 라이브러리


```python
import seaborn as sns
```

## barplot
* 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 기본적인 바 차트를 생성
  + 평균은 막대의 높이로, 편차는 에러바(error bar)로 표시

```python
plt.figure(figsize=(10,7)) 
sns.barplot(x='지역', y='대학수', data=dfg)
plt.title('지역별 대학 현황(컴퓨터관련학과)')
plt.xticks(rotation=90)
plt.show()
```

## Countplot
* 항목별 갯수를 세어주는 그래프

```python
area = sorted(list(set(df['지역'])))
area
```

```python
plt.figure(figsize=(10,7)) 
area = sorted(list(set(df['지역'])))

sns.countplot(x='지역', data=df, order = area).set_title('지역별 대학 현황(컴퓨터관련학과)')
plt.show()
```

```python
plt.figure(figsize=(10,7)) 
area = sorted(list(set(df['지역'])))

sns.countplot(x='지역', hue='설립유형', data=df, order = area).set_title('지역별 대학 현황(컴퓨터관련학과)')
plt.show()
```

```python
plt.figure(figsize=(10,7)) 
area = sorted(list(set(df['지역'])))

sns.countplot(y='지역', hue='설립유형', data=df, order = area).set_title('지역별 대학 현황(컴퓨터관련학과)')
plt.show()
```

## boxplot
* 다섯 수치 요약(five number summary)을 가지고 표현하는 그래프  
 + 최솟값 : 제 1사분위에서 1.5 IQR1을 뺀 위치 
 + 제 1사분위(Q1) : 25%의 위치를 의미 
 + 제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미 
 + 제 3사분위(Q3) : 75%의 위치를 의미 
 + 최댓값 : 제 3사분위에서 1.5 IQR을 더한 위치 
 + 이상치(Outlier) : 최솟값과 최댓값을 넘어가는 위치에 있는 값

```python
sns.boxplot(data=pnu[['경쟁률']])
plt.show()
```
